Configuración inicial y carga de librerías

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# Configuración de estilos para visualizaciones
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
%matplotlib inline
```

Carga y exploración de datos

```python
print("CARGANDO DATOS...")
print("=" * 50)

# Cargar dataset desde Google Drive o subir archivo
from google.colab import files
from io import BytesIO

# Intentar cargar el archivo train.csv
try:
    # Si estás en Google Colab, sube el archivo
    uploaded = files.upload()
    file_name = list(uploaded.keys())[0]
    datos = pd.read_csv(BytesIO(uploaded[file_name]))
    print(f"Archivo '{file_name}' cargado exitosamente.")
except:
    # Si no se puede cargar, usar dataset alternativo
    print("No se pudo cargar el archivo, usando dataset de respaldo...")
    # Dataset alternativo (puedes reemplazar esta URL con la correcta)
    url = "https://raw.githubusercontent.com/kabure/kaggle-mobile-price-classification/master/data/train.csv"
    datos = pd.read_csv(url)

# Información básica del dataset
print(f"Dimensiones del dataset: {datos.shape}")
print(f"\nNombres de columnas: {list(datos.columns)}")
print(f"\nTipos de datos:\n{datos.dtypes}")

# Verificar si la columna price_range existe
if 'price_range' not in datos.columns:
    # Si no existe, buscar una columna similar
    target_col = None
    for col in datos.columns:
        if 'price' in col.lower() or 'range' in col.lower():
            target_col = col
            break

    if target_col:
        datos = datos.rename(columns={target_col: 'price_range'})
        print(f"Renombrada columna '{target_col}' a 'price_range'")
    else:
        # Si no encontramos una columna objetivo, crear una ficticia para demostración
        print("No se encontró columna 'price_range'. Creando columna ficticia para demostración.")
        np.random.seed(42)
        datos['price_range'] = np.random.randint(0, 4, datos.shape[0])

# Mostrar primeras filas
print("\nPrimeras 10 filas del dataset:")
display(datos.head(10))
```

Análisis exploratorio y visualizaciones

```python
print("\nANÁLISIS EXPLORATORIO DE DATOS")
print("=" * 50)

# 1. Estadísticas descriptivas
print("Estadísticas descriptivas de variables numéricas:")
display(datos.describe())

# 2. Distribución de la variable objetivo
print("\nDistribución de la variable objetivo (rango de precios):")
distribucion_precios = datos['price_range'].value_counts()
print(distribucion_precios)

plt.figure(figsize=(10, 6))
sns.countplot(x='price_range', data=datos, palette='viridis')
plt.title('Distribución de Rangos de Precio de Móviles', fontsize=16, fontweight='bold')
plt.xlabel('Categoría de Precio (0: bajo, 1: medio, 2: alto, 3: muy alto)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()

# 3. Matriz de correlación
print("\nMatriz de correlación entre variables numéricas:")
correlaciones = datos.corr(numeric_only=True)

plt.figure(figsize=(16, 12))
sns.heatmap(correlaciones, annot=True, cmap='RdBu_r', center=0,
            square=True, fmt='.2f', linewidths=0.5)
plt.title('Mapa de Calor de Correlaciones entre Variables', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# 4. Análisis de variables específicas
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Relación entre RAM y precio
sns.boxplot(x='price_range', y='ram', data=datos, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Relación entre RAM y Rango de Precio')
axes[0, 0].set_xlabel('Rango de Precio')
axes[0, 0].set_ylabel('RAM (MB)')

# Relación entre batería y precio
if 'battery_power' in datos.columns:
    sns.boxplot(x='price_range', y='battery_power', data=datos, ax=axes[0, 1], palette='viridis')
    axes[0, 1].set_title('Relación entre Batería y Rango de Precio')
    axes[0, 1].set_xlabel('Rango de Precio')
    axes[0, 1].set_ylabel('Capacidad de Batería (mAh)')

# Relación entre cámara principal y precio
if 'pc' in datos.columns:
    sns.boxplot(x='price_range', y='pc', data=datos, ax=axes[1, 0], palette='viridis')
    axes[1, 0].set_title('Relación entre Cámara Principal y Rango de Precio')
    axes[1, 0].set_xlabel('Rango de Precio')
    axes[1, 0].set_ylabel('Megapíxeles de Cámara Principal')

# Relación entre tamaño de pantalla y precio
if 'sc_w' in datos.columns and 'sc_h' in datos.columns:
    datos['screen_size'] = datos['sc_w'] * datos['sc_h']  # Crear variable de tamaño de pantalla
    sns.boxplot(x='price_range', y='screen_size', data=datos, ax=axes[1, 1], palette='viridis')
    axes[1, 1].set_title('Relación entre Tamaño de Pantalla y Rango de Precio')
    axes[1, 1].set_xlabel('Rango de Precio')
    axes[1, 1].set_ylabel('Área de Pantalla (cm²)')

plt.tight_layout()
plt.show()
```

Reprocesamiento de datos

```python
print("\nPREPROCESAMIENTO DE DATOS")
print("=" * 50)

# Separar características y variable objetivo
X = datos.drop('price_range', axis=1)
y = datos['price_range']

# División en conjuntos de entrenamiento y prueba
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y)

print(f"Tamaño del conjunto de entrenamiento: {X_entrenamiento.shape}")
print(f"Tamaño del conjunto de prueba: {X_prueba.shape}")

# Estandarización de características
escalador = StandardScaler()
X_entrenamiento_esc = escalador.fit_transform(X_entrenamiento)
X_prueba_esc = escalador.transform(X_prueba)

print("\nPrimeras 5 filas de datos estandarizados:")
display(pd.DataFrame(X_entrenamiento_esc, columns=X.columns).head())
```

Implementación de máquina de vectores de soporte (SVM)

```python
print("\nIMPLEMENTACIÓN DE MÁQUINA DE VECTORES DE SOPORTE (SVM)")
print("=" * 50)

# Búsqueda de hiperparámetros óptimos
parametros_svm = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'linear', 'poly']
}

modelo_svm = SVC(random_state=42)
busqueda_svm = GridSearchCV(modelo_svm, parametros_svm, cv=5,
                           scoring='accuracy', n_jobs=-1, verbose=1)
busqueda_svm.fit(X_entrenamiento_esc, y_entrenamiento)

print(f"\nMejores parámetros para SVM: {busqueda_svm.best_params_}")
print(f"Mejor puntuación de validación cruzada: {busqueda_svm.best_score_:.4f}")

# Entrenamiento del modelo con mejores parámetros
mejor_svm = busqueda_svm.best_estimator_
predicciones_svm = mejor_svm.predict(X_prueba_esc)

# Evaluación del modelo SVM
precision_svm = accuracy_score(y_prueba, predicciones_svm)
puntuacion_f1_svm = f1_score(y_prueba, predicciones_svm, average='weighted')

print(f"\nPrecisión del modelo SVM: {precision_svm:.4f}")
print(f"Puntuación F1 del modelo SVM: {puntuacion_f1_svm:.4f}")

print("\nReporte de clasificación - SVM:")
print(classification_report(y_prueba, predicciones_svm))

# Matriz de confusión para SVM
plt.figure(figsize=(8, 6))
matriz_confusion = confusion_matrix(y_prueba, predicciones_svm)
sns.heatmap(matriz_confusion, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Bajo', 'Medio', 'Alto', 'Muy Alto'],
            yticklabels=['Bajo', 'Medio', 'Alto', 'Muy Alto'])
plt.title('Matriz de Confusión - Modelo SVM', fontsize=14, fontweight='bold')
plt.ylabel('Etiqueta Real')
plt.xlabel('Etiqueta Predicha')
plt.show()
```

Implementación de red neuronal

```python
print("\nIMPLEMENTACIÓN DE RED NEURONAL")
print("=" * 50)

# Definición de arquitecturas a probar
arquitecturas = [
    (100,),  # Una capa con 100 neuronas
    (200, 100),  # Dos capas: 200 y 100 neuronas
    (200, 100, 50),  # Tres capas: 200, 100 y 50 neuronas
    (300, 200, 100)  # Tres capas: 300, 200 y 100 neuronas
]

resultados_rn = {}

for i, capas in enumerate(arquitecturas):
    print(f"\nEntrenando red neuronal con arquitectura {capas}...")

    modelo_rn = MLPClassifier(
        hidden_layer_sizes=capas,
        activation='relu',
        solver='adam',
        alpha=0.001,
        batch_size='auto',
        learning_rate='constant',
        learning_rate_init=0.001,
        max_iter=500,
        random_state=42,
        early_stopping=True,
        validation_fraction=0.1,
        n_iter_no_change=10,
        verbose=False
    )

    modelo_rn.fit(X_entrenamiento_esc, y_entrenamiento)

    # Predicciones y evaluación
    predicciones_rn = modelo_rn.predict(X_prueba_esc)
    precision_rn = accuracy_score(y_prueba, predicciones_rn)
    puntuacion_f1_rn = f1_score(y_prueba, predicciones_rn, average='weighted')

    resultados_rn[f'Arquitectura_{i+1}'] = {
        'arquitectura': capas,
        'precision': precision_rn,
        'f1_score': puntuacion_f1_rn,
        'modelo': modelo_rn
    }

    print(f"Precisión: {precision_rn:.4f}, F1-score: {puntuacion_f1_rn:.4f}")

# Seleccionar el mejor modelo de red neuronal
mejor_rn = None
mejor_precision = 0
for clave, resultado in resultados_rn.items():
    if resultado['precision'] > mejor_precision:
        mejor_precision = resultado['precision']
        mejor_rn = resultado

print(f"\nMejor arquitectura: {mejor_rn['arquitectura']}")
print(f"Precisión del mejor modelo de red neuronal: {mejor_rn['precision']:.4f}")
print(f"Puntuación F1 del mejor modelo: {mejor_rn['f1_score']:.4f}")

# Reporte de clasificación del mejor modelo RN
predicciones_mejor_rn = mejor_rn['modelo'].predict(X_prueba_esc)
print("\nReporte de clasificación - Mejor Red Neuronal:")
print(classification_report(y_prueba, predicciones_mejor_rn))

# Matriz de confusión para la mejor red neuronal
plt.figure(figsize=(8, 6))
matriz_confusion_rn = confusion_matrix(y_prueba, predicciones_mejor_rn)
sns.heatmap(matriz_confusion_rn, annot=True, fmt='d', cmap='Greens',
            xticklabels=['Bajo', 'Medio', 'Alto', 'Muy Alto'],
            yticklabels=['Bajo', 'Medio', 'Alto', 'Muy Alto'])
plt.title('Matriz de Confusión - Mejor Modelo de Red Neuronal', fontsize=14, fontweight='bold')
plt.ylabel('Etiqueta Real')
plt.xlabel('Etiqueta Predicha')
plt.show()
```

Comparación de modelos

```python
print("\nCOMPARACIÓN DE MODELOS")
print("=" * 50)

# Crear DataFrame comparativo
comparacion = pd.DataFrame({
    'Modelo': ['SVM', 'Red Neuronal'],
    'Precisión': [precision_svm, mejor_rn['precision']],
    'F1-Score': [puntuacion_f1_svm, mejor_rn['f1_score']]
})

print("Comparación de rendimiento entre modelos:")
display(comparacion)

# Visualización comparativa
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Gráfico de precisión
ax[0].bar(comparacion['Modelo'], comparacion['Precisión'], color=['skyblue', 'lightgreen'])
ax[0].set_title('Comparación de Precisión entre Modelos', fontsize=14, fontweight='bold')
ax[0].set_ylabel('Precisión')
ax[0].set_ylim(0, 1)
for i, v in enumerate(comparacion['Precisión']):
    ax[0].text(i, v + 0.01, f'{v:.3f}', ha='center', fontweight='bold')

# Gráfico de F1-score
ax[1].bar(comparacion['Modelo'], comparacion['F1-Score'], color=['skyblue', 'lightgreen'])
ax[1].set_title('Comparación de F1-Score entre Modelos', fontsize=14, fontweight='bold')
ax[1].set_ylabel('F1-Score')
ax[1].set_ylim(0, 1)
for i, v in enumerate(comparacion['F1-Score']):
    ax[1].text(i, v + 0.01, f'{v:.3f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```
